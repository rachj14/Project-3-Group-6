```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('dataset/Mental Illness Survey 1.csv')
```

```python
print(df)
```

```python
print(df.columns)
```

```python
sns.countplot(x='Education', hue='I identify as having a mental illness', data=df, order=[0,1])
plt.xlabel('Mental Illness')
plt.ylabel('Education')
plt.show()
```

```python
grouped_data = df.groupby(['Education', 'I identify as having a mental illness']).size().unstack(fill_value=0)
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Education')
plt.ylabel('Mental Illness')
plt.title('Affect of Education on Mental Illness')
plt.legend(title='Mental Illness')
plt.show()
```

```python
sns.countplot(x='I identify as having a mental illness', hue='Education', data=df)
plt.xlabel('Categorical String Column')
plt.ylabel('Count')
plt.title('Count of 0s and 1s by Category')
plt.legend(title='Mental Illness')
plt.show()
```

```python
unique_categories = df['Education'].unique()

print(unique_categories)
```

```python
plt.hist(df['Education'], bins= len(unique_categories), color='skyblue', edgecolor='black')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.grid(True)
plt.show()
```

```python
contingency_table = pandas.crosstab(df['I have a gap in my resume'], df['I identify as having a mental illness'])
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='coolwarm')
plt.xlabel('Resume Gap')
plt.ylabel('I identify as having a mental illness')
plt.title('Contingency Table Heatmap')
plt.show()
```

```python
contingency_table = pandas.crosstab(df['I identify as having a mental illness'], df['I live with my parents'])
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='coolwarm')
plt.xlabel('I live with my parents')
plt.ylabel('I identify as having a mental illness')
plt.title('Contingency Table Heatmap')
plt.show()
```

```python
contingency_table = pandas.crosstab(df['I am unemployed'], df['I identify as having a mental illness'])
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='coolwarm')
plt.xlabel('I am unemployed')
plt.ylabel('I identify as having a mental illness')
plt.title('Contingency Table Heatmap')
plt.show()
```
